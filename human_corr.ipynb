```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster import hierarchy
```

```python
final = pd.read_excel('./final.xlsx')
```

```python
final_corr = final.corr()
```

```python
# correlation_matrix = pd.read_excel('./human_corr.xlsx')
correlation_matrix = final_corr
```

```python
# correlation_matrix = correlation_matrix.iloc[:,1:]
```

```python
Z = linkage(correlation_matrix, 'ward')
dendro = hierarchy.dendrogram(Z, no_plot=True)
order = dendro['leaves']
sorted_corr_matrix = correlation_matrix.iloc[order, order]
```

```python
plt.figure(figsize=(40, 30))
sns.heatmap(sorted_corr_matrix, annot=False, fmt=".2f", cmap='coolwarm',
            xticklabels=sorted_corr_matrix.columns, yticklabels=sorted_corr_matrix.columns)
plt.title('Heatmap of Variable Correlations After Reordering')
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.savefig('pic.jpg')
plt.show()
```
